```python
import math 
import matplotlib.pyplot as plt
math.sqrt(4)
```

```python
Bandwith=[0,20,20,20,20,20]
Hz=[0,220,440,880,1760,3520]
resolution=10
gain=[0,dfs.iloc[0]['220'],dfs.iloc[0]['440'],dfs.iloc[0]['880'],dfs.iloc[0]['1760'],dfs.iloc[0]['2520']]
```

```python
graph=np.zeros((len(Hz))*resolution)
xaxis=np.zeros((len(Hz))*resolution)
for i in range(len(Hz)):
    if gain[i]==0:
        continue
    k=Bandwith[i]
    y=math.fabs(gain[i]+9.8)
    x=math.log(Hz[i]/110,2)*resolution
    c=math.sqrt(math.fabs(k*(2*math.log(math.fabs(y*k*math.sqrt(2*math.pi)),math.e))))-x
    #print("k=",k," c=",c)
    #print("x=",x," y=",y)
    for j in range(len(graph)):
        thismouth=(1/(k*math.sqrt(2*math.pi)))*(math.pow(math.e,(-1/2*((math.pow(j+c-15,2)/k)))))
        graph[j]=graph[j]-thismouth
for j in range(len(graph)):
    xaxis[j]=math.pow(2,j/resolution)*110
plt.plot(xaxis[:53],graph[:53])
plt.show()
```
