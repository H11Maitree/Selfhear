```python
queue=[]
class queuecell:
    def __init__(self, positionx,positiony,positionz,gx,gy,gz):
        self.x = positionx
        self.y = positiony
        self.z = positionz
        self.distancepow2= ((positionx-gx)**2)+((positiony-gy)**2)+((positionz-gz)**2)
    def isthissmaller(self,cell2):
        if(self.distancepow2<cell2.distancepow2):
            return True
        else:
            return False
```

```python
def inserttoqueue(queuecell):
    if len(queue)==0:
        queue.append(queuecell)
        return 0
    else:
        for i in range(len(queue)):
            #print("queue",len(queue),i,queue[i])
            #print(queue[i].distancepow2,queuecell.distancepow2)
            if(queuecell.isthissmaller(queue[i])):
                queue.insert(i,queuecell)
                return 0
    queue.append(queuecell)
```

```python
#z y x
empty = np.zeros(shape=(10,100,100))
for iz in range(empty.shape[0]):
    print(iz)
    for iy in range(empty.shape[1]):
        for ix in range(empty.shape[2]):
            inserttoqueue(queuecell(ix,iy,iz,20,20,2))
```
